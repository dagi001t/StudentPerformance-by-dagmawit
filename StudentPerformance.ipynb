```python
# Dagmawit Tsegaye    1895/12
#This data set consists of the marks secured by the students in various subjects      
#this data analysis project is build using three different datasets and data format which is based on students and their performance 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



#Wrangle data
json_file = 'score_updated.json'
csv_file = 'StudentMentalhealth.csv'
excel_file = 'StudentsPerformance.xlsx'



# Load data into DataFrames
df_json = pd.read_json(json_file)
df_csv = pd.read_csv(csv_file)
df_excel = pd.read_excel(excel_file)   
```

```python
#Merging three different format datasets
df_combined = pd.concat([df_json, df_csv, df_excel], ignore_index=True) 
print(df_combined.head()) 
```

```python
# Assess data
# Display basic information about the DataFrame
print("Data Structure and Types:")
print(df_combined.info())


# Check for missing values
print("\nMissing Values:")
print(df_combined.isnull().sum())


# Check summary statistics for numerical columns
print("\nSummary Statistics:")
print(df_combined.describe())
```

```python
#Data Cleaning concept 
#Handling Categorical Data
# Convert categorical columns to numerical using one-hot encoding
df_student_data_encoded = pd.get_dummies(df_combined, columns=['gender', 'parental level of education', 'lunch', 'test preparation course'])
```

```python
#Showing Relatoinship
# Read the JSON file into a DataFrame
excel_file = 'StudentsPerformance.xlsx'
df_student_data = pd.read_excel(excel_file)

# Visualize Scores for Multiple Data Points
plt.figure(figsize=(15, 5))

# Bar plot for math, reading, and writing scores
plt.subplot(1, 3, 1)
sns.barplot(x=df_student_data.columns[5:], y=df_student_data.iloc[:, 5:].mean())
plt.title('Average Scores Across Students')
plt.ylabel('Average Score')

# Violin plot for math, reading, and writing scores
plt.subplot(1, 3, 2)
sns.violinplot(data=df_student_data.iloc[:, 5:])
plt.title('Distribution of Scores')

# Box plot for math, reading, and writing scores
plt.subplot(1, 3, 3)
sns.boxplot(data=df_student_data.iloc[:, 5:])
plt.title('Score Distribution')

plt.tight_layout()
plt.show()
```

```python
#Drawing Conclusions
mean_scores = df_combined[['math score', 'reading score', 'writing score']].mean()
median_scores = df_student_data[['math score', 'reading score', 'writing score']].median()
std_dev_scores = df_student_data[['math score', 'reading score', 'writing score']].std()

print("Mean Scores:")
print(mean_scores)
print("\nMedian Scores:")
print(median_scores)
print("\nStandard Deviation of Scores:")
print(std_dev_scores)
```

```python
#Communicating the results
plt.figure(figsize=(12, 6))

# Box plot for math, reading, and writing scores
sns.boxplot(data=df_combined[['math score', 'reading score', 'writing score','Hours']])
plt.title('Distribution of Scores Across Subjects')
plt.xlabel('Subjects and reading hour')
plt.ylabel('Score')
plt.show()    
```

```python
#Communicating the Results
sns.histplot(df_combined['gender'], kde=True)
plt.title('Distribution of Scores')
plt.show()

sns.boxplot(x='math score', y='gender', data=df_combined)
plt.title('Score Distribution by Subject')
plt.show() 


print("Data analysis completed successfully!  Prepared By:Dagmawit Tsegaye ")
```
